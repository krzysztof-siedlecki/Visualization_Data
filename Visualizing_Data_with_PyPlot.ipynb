```python
from matplotlib import pyplot as plt
```

```python
#Generating random data
import random
x = [i for i in range(1,101)]
y = []
for i in range(100):
  y.append(random.randint(1,101))
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Some Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
# Colors: w r y g k c b m
# Shapes: o p s d h x v . , 1 2 3 4 8 - + | : < > -- -.
plt.plot(x,y,'c+')
plt.show()
```

```python
lines = plt.plot(x,y)
plt.setp(lines, linewidth=3, color='y')
plt.show()
```

```python
y_1 = []
y_2 = []
for i in range(100):
  y_1.append(random.randint(1,101))
  y_2.append(random.randint(1,101))
```

```python
plt.plot(x,y_1,'gx',x,y_2,'ro')
plt.show()
```

```python
plt.plot(x,y_1,'g',x,y_2,'r')
plt.title("title")
plt.xlabel('red')
plt.ylabel('green')
plt.show()
```

```python
#plt.subplot(row,col,index)
plt.subplot(2,1,1)
plt.title("Graph 1")
plt.plot(x,y_1,'r')
plt.subplot(2,1,2)
plt.title('Graph 2')
plt.plot(x,y_2,'g')
plt.suptitle('Graphs')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title("Graph 1")
plt.plot(x,y_1,'r')
plt.subplot(1,2,2)
plt.title('Graph 2')
plt.plot(x,y_2,'g')
plt.suptitle('Graphs')
plt.show()
```
